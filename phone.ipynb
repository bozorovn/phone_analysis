```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('phone.csv', delimiter = ';')

df.head()
```

```python
df['ram_size'] = df['ram'].str.extract(r'(\d+) GB')
```

```python
df['ram_size']
```

```python
df[['ram_size_1', 'ram_size_2']] = df['ram'].str.extractall(r'(\d+) GB').unstack()
```

```python
df[['battery_capacity', 'charging_speed']] = df['battery'].str.extractall(r'(\d+) mAh.*?(\d+)W').unstack()

```

```python
df[['screen_size', 'resolution', 'refresh_rate']] = df['display'].str.extract(r'([\d.]+) inches, (\d+ x \d+) px, (\d+) Hz')
```

```python
df.info()
```

```python
df[['rear_camera', 'front_camera']] = df['camera'].str.extract(r'(\d+ MP(?: \+ \d+ MP)*)(?: Triple)? Rear & (\d+ MP) Front Camera')
```

```python
df['rear_camera'].isnull().sum()
```

```python
df[['rear_camera', 'front_camera']] = df['camera'].str.extract(r'(\d+ MP(?: \+ \d+ MP)*)?(?: Triple)? Rear(?: \& (\d+ MP(?: \+ \d+ MP)*))? Front Camera')
```

```python
df.head(7)
```

```python
df[['brand', 'model']] = df['model'].str.extract(r'(\w+)(?:\s+(.+))?')
```

```python
cleaned_phone = pd.DataFrame({
    'ram_size':df['ram_size'],
    'storage_size':df['ram_size_2'],
    'battery_capacity':df['battery_capacity'],
    'fast_charging':df['charging_speed'],
    'front_camera':df['front_camera']
})
```

```python
df['battery'].head(50)
```

```python
cleaned_phone.info()
```

```python

cleaned_phone['ram_size'] = pd.to_numeric(cleaned_phone['ram_size'].str.extract('(\d+)')[0], errors='coerce')
cleaned_phone['storage_size'] = pd.to_numeric(cleaned_phone['storage_size'].str.extract('(\d+)')[0], errors='coerce')
cleaned_phone['battery_capacity'] = pd.to_numeric(cleaned_phone['battery_capacity'].str.extract('(\d+)')[0], errors='coerce')
cleaned_phone['fast_charging'] = pd.to_numeric(cleaned_phone['fast_charging'], errors='coerce')  # Assuming 'fast_charging' is already numeric
cleaned_phone['front_camera'] = pd.to_numeric(cleaned_phone['front_camera'].str.extract('(\d+)')[0], errors='coerce')

```

```python
#finding mean values of columns
mean_storage_size = cleaned_phone['storage_size'].mean()
mean_battery_capacity = cleaned_phone['battery_capacity'].mean()
mean_front_camera = cleaned_phone['front_camera'].mean()
```

```python
#filling null values
cleaned_phone['storage_size'].fillna(mean_storage_size, inplace = True)
cleaned_phone['battery_capacity'].fillna(mean_battery_capacity, inplace = True)
cleaned_phone['fast_charging'].fillna(0, inplace = True)
cleaned_phone['front_camera'].fillna(mean_front_camera, inplace = True)
```

```python
cleaned_phone.info()
```

our data has become clean like a crystal

```python
cleaned_phone.to_csv('cleaned_phone_pandas.csv', index = False)
```
